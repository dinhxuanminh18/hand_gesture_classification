```python
import csv

import numpy as np
import tensorflow as tf

import sklearn as sk


data = 'model/asl_classification/asl_data.csv'
num_class = 36
classes = 36


x_dataset = np.loadtxt(data, delimiter = ',', dtype = np.float32, usecols = list(range(1, (21 * 2 + 1))))
y_dataset = np.loadtxt(data, delimiter = ',', dtype = np.float32, usecols = 0)

x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(x_dataset, y_dataset, test_size = 0.25, random_state = 42)
```

```python
model = tf.keras.models.Sequential(
   [
        tf.keras.layers.Input((21 * 2, )),
        tf.keras.layers.Dense(150, activation = 'relu'),
        tf.keras.layers.Dropout(0.4),
        tf.keras.layers.Dense(100, activation = 'relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(70, activation = 'relu'),
        tf.keras.layers.Dense(35, activation = 'softmax')
   ]
)

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.summary()

model.fit(
    x_train,
    y_train,
    epochs = 500,
    batch_size = 128,
    validation_data = (x_test, y_test)
)
```

```python
print(model.evaluate(x_test, y_test))
```

```python
print(model.evaluate(x_test, y_test))
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tf_lite_model = converter.convert()
with open('model/asl_classification/asl_classifier.tflite', 'wb') as f:
    f.write(tf_lite_model)
```
